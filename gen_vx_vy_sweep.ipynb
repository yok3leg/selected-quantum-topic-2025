```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 
#  list  volt   -200 um - 200 um
#  0.01 V = 1 um
# input -> [x1,x2,x3,...,xn]
# output -> [v1,v2,v3,....,vn]
#  n  step ,  x2-x1 = (xn-x1)/step

# for loop + np.append
ratio = 0.01
start_x = -200
stop_x = 200
start_y = -200
stop_y = 200
n = 50
row = 3

x = np.linspace(start_x,stop_x,n)
Vx= x*ratio
Vx = np.tile(Vx,row)
plt.plot(Vx)

y = np.linspace(start_y,stop_y,row)
Vy_temp= y*ratio
Vy = np.array([])
for i in range(np.size(y)*row):
  Vy = np.append(Vy,Vy_temp[i//row])
print(Vy)
plt.figure()
plt.plot(Vy)
```
