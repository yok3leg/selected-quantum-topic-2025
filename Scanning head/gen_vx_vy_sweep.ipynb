```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# โจทย์
# ต้องการ list ของ volt ที่ใช้ใส่ในหัวแสกน เพื่อใช้คุมตำแหน่งตั้งแต่ -200 um - 200 um
# โดยเรารู้ว่า 0.01 V = 1 um
# input -> [x1,x2,x3,...,xn]
# output -> [v1,v2,v3,....,vn]
# โดยที่ n คือจำนวน step การสแกน, เรารู้อีกอย่างว่า x2-x1 = (xn-x1)/step

# for loop + np.append
ratio = 0.01
start_x = -200
stop_x = 200
start_y = -200
stop_y = 200
n = 50
row = 3

x = np.linspace(start_x, stop_x, n)
Vx= x*ratio
Vx = np.tile(Vx,row)

Vx_reflected= (-1)*(Vx)
plt.plot(Vx, label='Vx')
plt.plot(Vx_reflected, label='-Vx')
plt.legend()
plt.show()

y = np.linspace(start_y,stop_y,row)
Vy_temp= y*ratio
Vy = np.array([])
for i in range(np.size(y)*row):
  Vy = np.append(Vy,Vy_temp[i//row])
print(Vy)
plt.figure()
plt.plot(Vy)



```

```python
x = np.linspace(start_x, stop_x, n)
Vx = x*ratio
Vx_reflected = (-1)*Vx

triangle_wave = np.array([])

for i in range(row) :
    if i%2 == 0:
        triangle_wave = np.append(triangle_wave, Vx)
    else :
        triangle_wave = np.append(triangle_wave, Vx_reflected)

plt.plot(triangle_wave, label='triangle_wave (Vx/ (-1)*Vx)')
plt.legend()
plt.show()

y = np.linspace(start_y,stop_y,row)
Vy= y*ratio
point= 3
Vy = np.repeat(Vy ,point)
plt.plot(Vy, drawstyle='steps-post', label='Vy stairs')
print(Vy)
plt.legend()
plt.show()



# num_triangles = 10
# x = np.linspace(start_x, stop_x, n)
# Vx = x*ratio
# Vx_reflected = (-1)*Vx

# triangle_wave = np.array([])
# for i in range(num_triangles):
#     triangle_wave = np.append(triangle_wave, Vx if i%2==0 else Vx_reflected)

# plt.plot(triangle_wave, label= f'{num_triangles} triangles(s)')
# plt.legend()
# plt.show()

```
